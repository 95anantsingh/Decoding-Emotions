```python
import os
import numpy as np
from torchstat import stat
from torchsummary import summary
from models import Dense, ProbingModel, ProbingDense
```

```python
dense_base = Dense('BASE',7,'cpu')
dense_large = Dense('LARGE',7,'cpu')
probing_base = ProbingModel('BASE',7)
probing_large = ProbingModel('LARGE',7)
probing_dense_base = ProbingDense('BASE',7,'cpu')
probing_dense_large = ProbingDense('LARGE',7,'cpu')
```

```python
base_layers = 12
large_layers = 24

seq_length = 256

base_feature_dim = 768
large_feature_dim = 1024

print('Sequence Length: 512')
print('Number of Labels: 7')
print('\n\nModel: dense_base')
print('Input shape:',(base_layers,seq_length,base_feature_dim))
summary(dense_base,(base_layers,seq_length,base_feature_dim))
print('\n\nModel: dense_large')
print('Input shape:',(large_layers,seq_length,large_feature_dim))
summary(dense_large,(large_layers,seq_length,large_feature_dim))

print('\n\nModel: probing_base')
print('Input shape:',(base_layers,seq_length,base_feature_dim))
summary(probing_base,(base_layers,seq_length,base_feature_dim))
print('\n\nModel: probing_large')
print('Input shape:',(large_layers,seq_length,large_feature_dim))
summary(probing_large,(large_layers,seq_length,large_feature_dim))

print('\n\nModel: probing_dense_base')
print('Input shape:',(base_layers,seq_length,base_feature_dim))
summary(probing_dense_base,(base_layers,seq_length,base_feature_dim))
print('\n\nModel: probing_dense_large')
print('Input shape:',(large_layers,seq_length,large_feature_dim))
summary(probing_dense_large,(large_layers,seq_length,large_feature_dim))
```
